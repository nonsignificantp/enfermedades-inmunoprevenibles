# Vaccine preventable diseases dataset

The data shows yearly reports of vaccine-preventable diseases from Buenos Aires city during the years 1995 to 2013. The original data was taken from [Buenos Aires' Statistics and Census Bureau website](http://www.estadisticaciudad.gob.ar/eyc/?p=28834).

```python
import numpy as np
import pandas as pd
```

Below is a copy-paste directly from the excel file downloaded form the website.

```python
year = "1995 1996 1997 1998 1999 2000 2001 2002 2003 2004 2005 2006 2007 2008 2009 2010 2011 2012 2013".split(' ')
```

```python
coqueluche = "96	46	28	123	52	29	28	16	35	19	3	2	42	26	12	13	116	94	121".replace(',', '').split('\t')
hepatitis_A = "928	1,493	1,281	1,145	1,350	1,245	694	789	1,107	816	369	146	35	27	21	28	11	20	7".replace(',', '').split('\t')
hepatitis_B = "263	371	544	381	291	311	238	288	265	97	74	74	55	47	50	55	60	68	50".replace(',', '').split('\t')
hepatitis_otras = "459	492	440	12	2	-	2	74	88	74	2	37	7	8	-	-	32	12	13".replace(',', '').split('\t')
paralisis_flaccida = "-	-	-	-	-	-	-	-	-	4	-	4	7	10	7	11	13	8	5".replace(',', '').split('\t')
parotiditis = "2,477	5,061	3,624	2,686	1,873	1,295	500	375	312	154	122	165	118	114	66	52	82	103	72".replace(',', '').split('\t')
rubeola = "1,892	989	833	1,632	967	765	180	86	41	4	1	6	1	88	2	-	2	1	1".replace(',', '').split('\t')
tetanos = "4	3	1	-	-	-	1	-	-	2	-	1	-	-	-	-	-	1	1".replace(',', '').split('\t')
```

Create a pandas dataframe and save it.

```python
vp_diseases = pd.DataFrame({'year':year,
                            'coqueluche':coqueluche,
                            'hepatitis_A':hepatitis_A,
                            'hepatitis_B':hepatitis_B,
                            'hepatitis_otras':hepatitis_otras,
                            'paralisis_flaccida':paralisis_flaccida,
                            'parotiditis':parotiditis,
                            'rubeola':rubeola,
                            'tetanos':tetanos})
```

```python
vp_diseases = vp_diseases.replace('-', '0')
vp_diseases = vp_diseases.astype(int)
```

```python
vp_diseases.to_csv('../datasets/vaccine-preventable-diseases.csv', index=False)
```
